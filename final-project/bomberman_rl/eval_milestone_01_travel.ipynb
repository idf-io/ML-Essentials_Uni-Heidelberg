```python
import json
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Manual settings
file_in = "results/eval_metrics.jsonl"
```

```python
df = pd.read_json(file_in, lines=True)
# df.set_index("match_name", inplace=True)

# INFO
if not baseline_agent in df.match_name.to_list():
    print("!!!!!",
          f"The baseline agent '{baseline_agent}' has not been found among the models",
          "!!!!!",
          sep="\n")

df['ID'] = df['match_name'].str[20:]
df.set_index('ID', inplace=True)

df
```

```python
# Filter for milestone
df1 = df[df['milestone'] == 1]
```

```python
df1.plot(kind='bar', y='coins_mean', yerr='coins_std', \
         color="#1f77b4", legend=False, figsize=(10,5))

plt.title('Mean coins collected per round', pad=20, size=15)
```

```python
df1.plot(kind='bar', y='invalid_moves', \
         color="#ff7f0e", legend=False, figsize=(10,5))

plt.title('Mean invalid moves collected per round', pad=20, size=15)
```

```python
df1.plot(kind='bar', y='steps_collect_coins_mean', yerr='steps_collect_coins_std', \
         color="#2ca02c", ylim=(0,160), legend=False, figsize=(10,5))

plt.title('Mean steps needed to collect all coins per round', pad=20, size=15)
```
